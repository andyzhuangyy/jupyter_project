```python
# visualization
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
import matplotlib.pyplot as plt
%matplotlib notebook
n_bins=100
x_max = 50
r_ndarray = ndarray[:,ndarray[0]<x_max]

t = r_ndarray[0,:]
x = r_ndarray[1,:]
y = r_ndarray[2,:]
z = r_ndarray[3,:]
u = r_ndarray[4,:]

fig, axs = plt.subplots(1, 2, tight_layout=True)
# N is the count in each bin, bins is the lower-limit of the bin
N, bins, patches = axs[0].hist(t, bins=n_bins)
# Now we format the y-axis to display percentage
axs[1].hist(t, bins=n_bins, density=True)
axs[1].yaxis.set_major_formatter(PercentFormatter(xmax=n_bins/x_max))

# fig1, axs1 = plt.subplots(1, 2, tight_layout=True)
# # N is the count in each bin, bins is the lower-limit of the bin
# N, bins, patches = axs1[0].hist(x, bins=n_bins)
# # Now we format the y-axis to display percentage
# axs1[1].hist(x, bins=n_bins, density=True)
# axs1[1].yaxis.set_major_formatter(PercentFormatter(xmax=n_bins))

# fig2, axs2 = plt.subplots(1, 2, tight_layout=True)
# N, bins, patches = axs2[0].hist(y, bins=n_bins)
# axs2[1].hist(y, bins=n_bins, density=True)
# axs2[1].yaxis.set_major_formatter(PercentFormatter(xmax=n_bins))

fig3, axs3 = plt.subplots(1, 2, tight_layout=True)
N, bins, patches = axs3[0].hist(z, bins=n_bins)
axs3[1].hist(z, bins=n_bins, density=True)
axs3[1].yaxis.set_major_formatter(PercentFormatter(xmax=n_bins))

fig, axs = plt.subplots(1, 2, tight_layout=True)
N, bins, patches = axs[0].hist(u, bins=n_bins)
axs[1].hist(u, bins=n_bins, density=True)
axs[1].yaxis.set_major_formatter(PercentFormatter(xmax=n_bins))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# the random data
x = ndarray[0,:]
y = ndarray[1,:]

# definitions for the axes
left, width = 0.1, 0.65
bottom, height = 0.1, 0.65
spacing = 0.005

rect_scatter = [left, bottom, width, height]
rect_histx = [left, bottom + height + spacing, width, 0.2]
rect_histy = [left + width + spacing, bottom, 0.2, height]

# start with a rectangular Figure
plt.figure(figsize=(8, 8))

ax_scatter = plt.axes(rect_scatter)
ax_scatter.tick_params(direction='in', top=True, right=True)
ax_histx = plt.axes(rect_histx)
ax_histx.tick_params(direction='in', labelbottom=False)
ax_histy = plt.axes(rect_histy)
ax_histy.tick_params(direction='in', labelleft=False)

# the scatter plot:
ax_scatter.scatter(x, y)

# now determine nice limits by hand:
binwidth = 0.25
lim = np.ceil(np.abs([x, y]).max() / binwidth) * binwidth
ax_scatter.set_xlim((-lim, lim))
ax_scatter.set_ylim((-lim, lim))

bins = np.arange(-lim, lim + binwidth, binwidth)
ax_histx.hist(x, bins=bins)
ax_histy.hist(y, bins=bins, orientation='horizontal')

ax_histx.set_xlim(ax_scatter.get_xlim())
ax_histy.set_ylim(ax_scatter.get_ylim())

plt.show()
```

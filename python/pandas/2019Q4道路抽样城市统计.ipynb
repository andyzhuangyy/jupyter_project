```python
def fun(adcode):
    if (adcode // 100) % 100 == 90:
        return adcode
    if adcode//10000 in [11, 12, 31, 50]:
        return adcode // 10000 * 10000
    else:
        return adcode // 100 * 100


import pandas as pd


columns=['link_id', 'grid_id', 'adcode', 'region', 'kind', 'length', 
         'lane', 'cond', 'light', 'eye', 'rev', 'adj', 'uflag']
path = '/Users/andyzhuang/Documents/0-业务/11-评测/01-道路数据评测/2019Q4/道路抽样/1000万公里道路抽样-非高速-20191129_75/4/links_in_grid_真实使用.mid'
df = pd.read_csv(path, delimiter=",", header=None, names=columns)
df = df[['link_id', 'adcode', 'length', 'lane', 'cond', 'light', 'eye', 'rev', 'adj']]
df['city_code'] = df.adcode.apply(fun)
df
```

```python
path_city = '/Users/andyzhuang/Documents/0-业务/参考文档/city_tab.csv'
col_city = ['adcode', 'city_name']
df_city = pd.read_csv(path_city, delimiter="\t", header=None, names=col_city)
df_city = df_city.query('adcode % 100 == 0 or ((adcode // 100) % 100 == 90)')
df_city['city_code'] = df_city.adcode.apply(fun)
# df_city = df_city.groupby('city_name').min()
df_city
```

```python
grouped = df.groupby('city_code')
# grouped.sum()
# df_sum = grouped.sum()[['city_name','length', 'lane', 'cond', 'light', 'eye', 'rev', 'adj']]
df_merge = pd.merge(grouped.sum(),df_city, how='left', on=['city_code'])[['city_code', 'city_name', 'length', 'lane', 'cond', 'light', 'eye', 'rev', 'adj']]
df_merge
df_sum=df_merge

df_sum.to_csv('stats_links_in_grid.csv')
df_sum
```
